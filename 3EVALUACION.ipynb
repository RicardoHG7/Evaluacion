## Desarrollar un análisis comparativo respecto a las 3 ciudades elegidas para el servicio Airbnb sobre las siguientes variables:

```python
#nuevas columnas llamadas "ciudad" a cada df
df_California['ciudad'] = 'EUA'
df_Mexico['ciudad'] = 'MEX'
df_Girona['ciudad'] = 'ESP'
```

```python
df_total = pd.concat([df_California, df_Mexico, df_Girona], ignore_index=True)
```

```python
df_total.info()
```

```python
from funpymodeling.exploratory import freq_tbl
#Obtengo un análisis univariado de las variables categóricas
freq_tbl(df_total)
```

```python
#Obtengo un análisis univariado de las variables categóricas
freq_tbl(df_California)
```

```python
freq_tbl(df_Girona)
```

```python
freq_tbl(df_Mexico)
```

```python
table_California= freq_tbl(df_California['host_response_time'])
table_Girona= freq_tbl(df_Girona['host_response_time'])
table_Mexico= freq_tbl(df_Mexico['host_response_time'])
```

```python
#Obtengo un análisis univariado de una variable categórica en específico
table_California
```

```python
table_Girona
```

```python
table_Mexico
```

```python
Filtro1= table_California.copy()
Filtro2= table_Girona.copy()
Filtro3= table_Mexico.copy()
```

```python
#Ajusto el indice de mi dataframe
Filtro_index1= Filtro1.set_index('host_response_time')
Filtro_index2= Filtro2.set_index('host_response_time')
Filtro_index3= Filtro3.set_index('host_response_time')
```

```python
#Gráfico de barras
fig, axs = plt.subplots(1, 3, figsize=(12,4), sharey=True)
plt.subplots_adjust(wspace=0.1)

Filtro_index1["percentage"].plot(kind='bar', ax=axs[0], color= 'blue', width=0.8)
Filtro_index2["percentage"].plot(kind='bar', ax=axs[1], color= 'orange', width=0.8)
Filtro_index3["percentage"].plot(kind='bar', ax=axs[2], color= 'red', width=0.8)
axs[0].set_title('California')
axs[1].set_title('Girona')
axs[2].set_title('Mexico')

for ax in axs:
    ax.grid(axis='y')
 
fig.text(0.08, 0.5, 'Porcentaje', va= 'center', rotation='vertical')
fig.suptitle('Tiempo que tarda un anfitrión en responder a una solicitud de reserva')
plt.show()
```

```python
# Concatenamos los DataFrames
Filtro_concat = pd.concat([Filtro_index1, Filtro_index2, Filtro_index3], axis=1, keys=['California', 'Girona', 'Mexico'])

porcentajes_Girona = Filtro_concat[('Girona', 'percentage')]
porcentajes_Mexico = Filtro_concat[('Mexico', 'percentage')]
porcentajes_California = Filtro_concat[('California', 'percentage')]

ax = pd.concat([porcentajes_California, porcentajes_Girona, porcentajes_Mexico], axis=1).plot(kind='bar', stacked=True, figsize=(10, 6), width=0.8)

plt.title('Tiempo de respuesta por país', fontsize=14)
plt.xlabel('host_response_time')
plt.ylabel('Porcentaje')

ax.grid(axis='y')

plt.show()
```

```python
#Obtengo un filtro de los valores más reelevantes de la variables categórica seleccionada
Filtro1_2= table_California1.copy()
Filtro2_2= table_Girona1.copy()
Filtro3_2= table_Mexico1.copy()
```

```python
#Ajusto el indice de mi dataframe
Filtro_index1_2= Filtro1_2.set_index('host_identity_verified')
Filtro_index2_2= Filtro2_2.set_index('host_identity_verified')
Filtro_index3_2= Filtro3_2.set_index('host_identity_verified')
```

```python

#Gráfico de barras
fig, axs = plt.subplots(1, 3, figsize=(12,4), sharey=True)
plt.subplots_adjust(wspace=0.1)

Filtro_index1_2["percentage"].plot(kind='bar', ax=axs[0], color= 'red', width=0.8)
Filtro_index2_2["percentage"].plot(kind='bar', ax=axs[1], color= 'orange', width=0.8)
Filtro_index3_2["percentage"].plot(kind='bar', ax=axs[2], color= 'green', width=0.8)
axs[0].set_title('California')
axs[1].set_title('Girona')
axs[2].set_title('Mexico')
for ax in axs:
    ax.grid(axis='y')
fig.text(0.08, 0.5, 'Porcentaje', va= 'center', rotation='vertical')
fig.suptitle('¿identidad del Host esta verificada?')
plt.show()
```

```python
# Concatenamos los DataFrames
Filtro_concat = pd.concat([Filtro_index1_2, Filtro_index2_2, Filtro_index3_2], axis=1, keys=['California', 'Girona', 'Mexico'])
porcentajes_California = Filtro_concat[('California', 'percentage')]
porcentajes_Girona = Filtro_concat[('Girona', 'percentage')]
porcentajes_Mexico = Filtro_concat[('Mexico', 'percentage')]
ax = pd.concat([porcentajes_California, porcentajes_Girona, porcentajes_Mexico], axis=1).plot(kind='bar', stacked=True, figsize=(10, 6), width=0.8)
plt.title("¿identidad del Host esta verificada por país?", fontsize=14)
plt.xlabel('Respuesta')
plt.ylabel('Porcentaje')
ax.grid(axis='y')
plt.show()
```

```python
df_grouped = df_total.groupby(['number_of_reviews', 'ciudad']).size().reset_index(name='conteo')
```

```python
df_grouped
```

```python
import matplotlib.pyplot as plt

# Filtramos los datos por ciudad
df_California = df_grouped[df_grouped['ciudad'] == 'EUA']
df_Mexico = df_grouped[df_grouped['ciudad'] == 'MEX']
df_Girona = df_grouped[df_grouped['ciudad'] == 'ESP']

# Creamos una figura con un tamaño de 10x6 pulgadas y 3 subgráficos (uno para cada ciudad)
plt.figure(figsize=(15, 5))

# Graficamos una línea para cada ciudad en su subgráfico correspondiente
plt.subplot(1, 3, 1)
plt.plot(df_California['number_of_reviews'], df_California['conteo'], label='EUA', color="blue")
plt.xlabel('Número de revisiones')
plt.ylabel('Conteo')
plt.title('EUA')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(df_Mexico['number_of_reviews'], df_Mexico['conteo'], label='MEX',color="green")
plt.xlabel('Número de revisiones')
plt.title('MEX')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(df_Girona['number_of_reviews'], df_Girona['conteo'], label='ESP',color="orange")
plt.xlabel('Número de revisiones')
plt.title('ESP')
plt.legend()

# Ajustamos el espaciado entre los subgráficos para evitar que se superpongan
plt.tight_layout()

# Mostramos el gráfico
plt.show()
```
